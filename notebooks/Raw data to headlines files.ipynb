```python
import sys
from outletsBiasRatings import outletsBiasRatingsAllSides,outletAbbreviationToFullName
import pandas as pd
import os
import re
```

```python
basePath = 'D:/media'
outletsBiasRatings = outletsBiasRatingsAllSides
outlets = list(outletsBiasRatings.keys())
startYear = 2000
endYear = 2019
```

```python
for outlet in outlets:
    print( outlet,end=',')
    inputPath=f'D:/media/{outlet}'
    outputPath=f'./headlinesData/{outlet}'
    if not os.path.exists(outputPath):
        os.makedirs(outputPath)
    files=os.listdir(inputPath)
    minimumNumberOfHeadlinesPerYearForInclusion=0
    for file in files:
        if re.match("^\d{4}.csv$", file):
            inputFilePath=os.path.join(inputPath,file)
            df=pd.read_csv(inputFilePath,usecols=['title','url'])
            if len(df['title'].dropna().values)>minimumNumberOfHeadlinesPerYearForInclusion:
                outputFilePath=os.path.join(outputPath,file)
                df.to_csv(outputFilePath,index=False)
```
