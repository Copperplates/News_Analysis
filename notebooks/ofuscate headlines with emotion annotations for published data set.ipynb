```python
import sys
from outletsBiasRatings import outletsBiasRatingsAllSides,outletAbbreviationToFullName
from matplotlib.pyplot import *
import pandas as pd
import os
import re
```

```python
basePath = 'D:/media'
outletsBiasRatings = outletsBiasRatingsAllSides
outlets = list(outletsBiasRatings.keys())
startYear = 2000
endYear = 2019
```

```python
def keepFirstAndLastWordInHeadline( headline):
    try:
        headlineWordsL = headline.split()
        obfuscatedHeadlineStr=' ... '.join([headlineWordsL[0],headlineWordsL[-1]])
    except:
        obfuscatedHeadlineStr=headline
        
    return obfuscatedHeadlineStr
```

```python
targetModel='headlinesDataWithEmotionLabelsAnnotationsFromDistilRoberta'
for outlet in outlets:
    print( outlet,end=',')
    inputPath=f'./{targetModel}/{outlet}'
    outputPath=f'D:/data/moral outrage headlines/published data set/{targetModel}/{outlet}'
    if not os.path.exists(outputPath):
        os.makedirs(outputPath)
    files=os.listdir(inputPath)
    for file in files:
        if re.match("^\d{4}.csv$", file):
            inputFilePath=os.path.join(inputPath,file)
            outputFilePath=os.path.join(outputPath,file)
            df=pd.read_csv(inputFilePath,usecols=['emotion','headline','url'])
            df['headline']=df['headline'].apply(keepFirstAndLastWordInHeadline)
            df.to_csv(outputFilePath,index = False)
```
