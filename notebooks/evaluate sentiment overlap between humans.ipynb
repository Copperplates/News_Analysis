```python
df=pd.read_csv('./otagoSentimentOverlapBetweenHumans.csv')

d={}
unequalsCounter=0
for i, row in df.iterrows():
    headline=row['headline'] 
    sentiment= row['sentiment']
    if headline not in d:
        d[headline]=[sentiment]
    else:
        d[headline].append(sentiment)
d
```

```python
c=0
for v in d.values():
    if v[0]==v[1]:
        c+=1
c/len(d.values())
```

```python
#Cohen’s Kappa
chanceP= 0.5
((c/len(d.values()))-chanceP)/(1-chanceP)
```
