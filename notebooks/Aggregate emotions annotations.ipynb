```python
import sys
from outletsBiasRatings import outletsBiasRatingsAllSides,outletAbbreviationToFullName
from matplotlib.pyplot import *
import matplotlib.pyplot as plt
import pandas as pd
import pickle
import os
import time
from joblib import Parallel, delayed
from transformers import AutoTokenizer, AutoModelWithLMHead
```

```python
outletsBiasRatings = outletsBiasRatingsAllSides
outlets = list(outletsBiasRatings.keys())
startYear = 2000
endYear = 2019
years=[str(year) for year in range(startYear,endYear+1)]
emotions=['anger','disgust','fear','joy','sadness', 'surprise','neutral'] #j-hartmann/emotion-english-distilroberta-base 
print(years)
print(len(outlets))
print(outlets)
```

```python
emotionsDictionary={}
for outlet in outlets:
    print( outlet , end=',')
    emotionsDictionary[outlet]={}
    inputPath=f'./headlinesDataWithEmotionLabelsAnnotationsFromDistilRoberta\{outlet}'
    for year in years:
        try:
            inputFilePath=os.path.join(inputPath,year+'.csv')
            df=pd.read_csv(inputFilePath)
        except: #no file
            continue

        emotionsDictionary[outlet][year]={}   
        for emotion in emotions:
            emotionsDictionary[outlet][year][emotion]=df[df['emotion']==emotion]['emotion'].count()/len(df)
```

## Save emotions labels annotations

```python
import pickle
with open('emotionsDictionary.pickle', 'wb') as handle:    
    pickle.dump(emotionsDictionary, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
