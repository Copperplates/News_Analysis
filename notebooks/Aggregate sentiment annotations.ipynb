```python
import sys
import os
from outletsBiasRatings import outletsBiasRatingsAllSides,outletAbbreviationToFullName
from matplotlib.pyplot import *
import pandas as pd
import pickle
import time
from joblib import Parallel, delayed
from transformers import AutoTokenizer, AutoModelWithLMHead
```

```python
outletsBiasRatings = outletsBiasRatingsAllSides
outlets = list(outletsBiasRatings.keys())
startYear = 2000
endYear = 2019
years=[str(year) for year in range(startYear,endYear+1)]

print(years)
print(len(outlets))
print(outlets)
```

```python
sentimentDictionary={}
for outlet in outlets:
    print( outlet , end=',')
    sentimentDictionary[outlet]={}
#     inputPath=rf'./headlinesDataWithSentimentLabelsAnnotationsFromDistilbertSST2/{outlet}'
    inputPath=rf'./headlinesDataWithSentimentLabelsAnnotationsFromSentimentRobertaLargeModel\{outlet}'
    for year in years:
        try:
            inputFilePath=os.path.join(inputPath,year+'.csv')
            df=pd.read_csv(inputFilePath)
        except: #no file
            continue
        sentimentDictionary[outlet][year]=df['sentiment'].astype(float).mean(skipna=True)        
```
