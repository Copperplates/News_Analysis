```python
from os import listdir
from os.path import isfile, join
import re
import csv
import numpy as np
import logging
import os
import pandas as pd 
import sys
import random
import math
import warnings
import dataframe_image as dfi
import matplotlib.pyplot as plt
import numpy as np

from outletsBiasRatings import outletsBiasRatingsAllSides, outletAbbreviationToFullName 

outletsBiasRatings = outletsBiasRatingsAllSides
outlets = list(outletsBiasRatings.keys())

startYear=2000
endYear = 2019
years = list(range(startYear,endYear+1))
print(len( outlets ))
print(outlets)
print(years)
```

```python
numberOfArticles = {}
for outlet in outlets:
    print( outlet, end=' ')
    numberOfArticles[outlet]={}
    filePath = os.path.normpath(f'./headlinesData/{outlet}') 
    for year in  years:
        yearFilePath = os.path.join(filePath, str(year)+'.csv') 
        try:
            df = pd.read_csv(yearFilePath)
        except: #no data
#             print(" exception raised by ", outlet, year)
            continue
        numberOfArticles[outlet][year]=df.shape[0] #Number of rows in data frame
```

```python
numberOfArticles
```

## Total number of  headlines per outlet for a specified time frame

```python
for outlet in numberOfArticles:
    print(outlet, sum(numberOfArticles[outlet].values()))
maxNumberOfArticles = max([sum(numberOfArticles[outlet].values()) for outlet in numberOfArticles])
outletWithMaxNumberOfArticles = outlets[[sum(numberOfArticles[outlet].values()) for outlet in numberOfArticles].index(maxNumberOfArticles)]
minNumberOfArticles = min([sum(numberOfArticles[outlet].values()) for outlet in numberOfArticles])
outletWithMinNumberOfArticles = outlets[[sum(numberOfArticles[outlet].values()) for outlet in numberOfArticles].index(minNumberOfArticles)]
averageNumberOfArticles = sum([sum(numberOfArticles[outlet].values()) for outlet in numberOfArticles])/len(numberOfArticles.keys())
print("max",maxNumberOfArticles,outletWithMaxNumberOfArticles)
print("min",minNumberOfArticles,outletWithMinNumberOfArticles)
print("average",averageNumberOfArticles)
```

## Total number of articles overall

```python
sum([sum(numberOfArticles[outlet].values()) for outlet in numberOfArticles])
```

## Total number of articles per year

```python
nHeadlinesList=[]
for year in years:
    nHeadlines=0
    for outlet in outlets:
        try:
            nHeadlines+=numberOfArticles[outlet][year]
        except:
            continue
    print( year, nHeadlines)
    nHeadlinesList.append(nHeadlines)
```
